```python
#import libraries
import pandas as pd
```

```python
# Function to analyse dataframe
def analysis(df):
    genre_df = df.groupby('book genre').sum().reset_index()
    genre_df['Return Rate'] = (genre_df['# of sold but returned books']/genre_df['# of all sold books'])*100
    genre_df['Sold with discount Rate'] = (genre_df['# of books sold with discount']/genre_df['# of all sold books'])*100
    genre_df[' % Sold to males'] = (genre_df['# of books sold to males']/genre_df['# of all sold books'])*100
    genre_df['% Sold to females'] = (genre_df['# of books sold to females']/genre_df['# of all sold books'])*100
    
    genre_and_seller_df = df.groupby(['seller', 'book genre']).sum().reset_index()
    genre_and_seller_df['Retrun Rate'] = (genre_and_seller_df['# of sold but returned books']/genre_and_seller_df['# of all sold books'])*100
    
    return genre_df, genre_and_seller_df
```

```python
df = pd.read_csv("data_01102021.csv")
genre_df, genre_and_seller_df = analysis(df)
```
